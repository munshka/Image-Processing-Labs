```python
import matplotlib.pyplot as plt
```

```python
img= data.cat()
plt.figure(figsize=(15,5),dpi=200) #size boro korchi image er

#side by side plotting an image is called subplot
#subplot(nroews,ncolumns,index)
#R-> [:,:,0] dimension 1
#G-> [:,:,1] dimension 2
#B-. [:,:,2] dimension 3

plt.subplot(1,3,1)
io.imshow(img[:,:,0])
plt.title("Red")

plt.subplot(1,3,2)
io.imshow(img[:,:,1])
plt.title("Green")

plt.subplot(1,3,3)
io.imshow(img[:,:,2])
plt.title("Blue")

plt.show()
```

```python
img= data.cat()
plt.imshow(img)

plt.title("Cat")
plt.xlabel("Width")
plt.ylabel("Heigth")

plt.show()
```

```python
img= data.cat()
plt.figure(figsize=(15,5),dpi=200) #size boro korchi image er

plt.subplot(1,2,1)
io.imshow(img[:,:,:])
plt.title("Original")

plt.subplot(1,2,2)
io.imshow(img[:,::-1,:])
plt.title("Flipped")

plt.show()
```
