```python
import numpy as np
from skimage import io, data, exposure
import matplotlib.pyplot as plt
```

```python
I= data.camera()
io.imshow(I)
```

```python
I.shape
```

```python
print(I)
```

```python
I.min()
```

```python
I.max()
```

```python
I.dtype #255=11111111 so if we add 1 it will become 0, so we will convert it to float
```

```python
I= I/255.0    #to convert we will divide by 255
```

```python
I.dtype
```

```python
I.max()
```

```python
hist, bins_hist = exposure.histogram(I, nbins=256, normalize=False) #qunatization level=256, normalize= exact count values
#bins_hist-> array 0 to 255
plt.plot(bins_hist, hist)
plt.xlabel("Intensity values")
plt.ylabel("# pixels")
plt.title("histogram")

plt.show()
```

```python
hist, bins_hist = exposure.histogram(I, nbins=256, normalize=True) #qunatization level=256, normalize= normalize y axis
plt.plot(bins_hist, hist)
plt.xlabel("Intensity values")
plt.ylabel("# pixels")
plt.title("histogram")

plt.show()
```

```python
I2 = I + 0.5 #means adding 0.5 intensity in 0-1, again means adding 100+ intensity in 0-255 range
io.imshow(I2)
```

```python
print(I2) # as some values are greater than 1 , it will give error to plot the values grater than 1, we will clip it
```

```python
I2= np.clip(I+0.5 , 0, 1) #the range is 0-1, if value<0 then values will become 0
                                           # if value>0 then values will become 1
io.imshow(I2)
```

```python
plt.subplot(1,2,1)
io.imshow(I)
plt.title("Original image")

plt.subplot(1,2,2)
io.imshow(I2)
plt.title("clipped")
```

```python
I3= np.clip(I-0.5 , 0, 1) #the range is 0-1, if value<0 then values will become 0
                                           # if value>0 then values will become 1
io.imshow(I3)
```

```python
plt.figure(figsize=(15,5),dpi=200) #size boro korchi image er

plt.subplot(1,3,1)
io.imshow(I)
plt.title("Original image")

plt.subplot(1,3,2)
io.imshow(I2)
plt.title("clipped")

plt.subplot(1,3,3)
io.imshow(I3)
plt.title("Clipped -0.5")

```

```python
A1= np.clip(I+0.1 , 0, 1)
A2= np.clip(I+0.2 , 0, 1)
A3= np.clip(I+0.3 , 0, 1)
A4= np.clip(I+0.4 , 0, 1)
```

```python
plt.subplot(1,4,1)
io.imshow(A1)
plt.title("I+0.1")

plt.subplot(1,4,2)
io.imshow(A2)
plt.title("I+0.2")

plt.subplot(1,4,3)
io.imshow(A3)
plt.title("I+0.3")

plt.subplot(1,4,4)
io.imshow(A4)
plt.title("I+0.4")

```

```python
plt.figure(figsize=(4,4),dpi=200) #size boro korchi image er
                                  #

plt.subplot(2,2,1)
io.imshow(A1)
plt.title("I+0.1")

plt.subplot(2,2,2)
io.imshow(A2)
plt.title("I+0.2")

plt.subplot(2,2,3)
io.imshow(A3)
plt.title("I+0.3")

plt.subplot(2,2,4)
io.imshow(A4)
plt.title("I+0.4")
```
